```python
energy.rename(columns = { 
                         'country_long':'Country',
                         'estimated_generation_gwh_2013':'2013',
                         'estimated_generation_gwh_2014':'2014',
                         'estimated_generation_gwh_2015':'2015',
                         'estimated_generation_gwh_2016':'2016',
                        'estimated_generation_gwh_2017':'2017'}, inplace = True)
```

```python
#change column attributes to numeric so a to be able to perform subtotals

energy["2013"] = pd.to_numeric(energy["2013"])
energy["2014"] = pd.to_numeric(energy["2014"])
energy["2015"] = pd.to_numeric(energy["2015"])
energy["2016"] = pd.to_numeric(energy["2016"])
energy["2017"] = pd.to_numeric(energy["2017"])


energy_subtotal2013 = energy.stb.freq(['Country'], value='2013')
energy_subtotal2014 = energy.stb.freq(['Country'], value='2014')
energy_subtotal2015 = energy.stb.freq(['Country'], value='2015')
energy_subtotal2016 = energy.stb.freq(['Country'], value='2016')
energy_subtotal2017 = energy.stb.freq(['Country'], value='2017')

print(energy_subtotal2013)
print('')
print(energy_subtotal2014)
print('')
print(energy_subtotal2015)
print('')
print(energy_subtotal2016)
print('')
print(energy_subtotal2017)
```

```python

angola_energy= pd.DataFrame(energy[(energy['Country'] == 'Angola')])

congo_energy =pd.DataFrame(energy[(energy['Country'] == 'Congo')])


central_africa_energy = angola_energy.append([congo_energy])

central_africa_energy["2013"] = pd.to_numeric(central_africa_energy["2013"])
central_africa_energy["2014"] = pd.to_numeric(central_africa_energy["2014"])
central_africa_energy["2015"] = pd.to_numeric(central_africa_energy["2015"])
central_africa_energy["2016"] = pd.to_numeric(central_africa_energy["2016"])
central_africa_energy["2017"] = pd.to_numeric(central_africa_energy["2017"])

central_africa_energy_subtotal2013 = central_africa_energy.stb.freq(['Country'],value='2013')
central_africa_energy_subtotal2014 = central_africa_energy.stb.freq(['Country'],value='2014')
central_africa_energy_subtotal2015 = central_africa_energy.stb.freq(['Country'],value='2015')
central_africa_energy_subtotal2016 = central_africa_energy.stb.freq(['Country'],value='2016')
central_africa_energy_subtotal2017 = central_africa_energy.stb.freq(['Country'],value='2017')




central_africa_energy_subtotal1 =central_africa_energy_subtotal2013.merge(central_africa_energy_subtotal2014, on='Country').merge(central_africa_energy_subtotal2015,on= 'Country').merge(central_africa_energy_subtotal2016,on= 'Country').merge(central_africa_energy_subtotal2017,on= 'Country')



central_africa_energy_subtotal1 = central_africa_energy_subtotal1.drop('percent_x',axis=1).drop('cumulative_2013',axis=1).drop('cumulative_percent_x', axis=1).drop('percent_y',axis=1).drop('cumulative_2014',axis=1).drop('cumulative_percent_y',axis=1).drop('cumulative_2015',axis=1).drop('cumulative_2016',axis=1).drop('cumulative_2017',axis=1).drop('cumulative_percent',axis=1).drop('percent',axis=1)



central_africa_energy_subtotal1[['Country', '2013','2014','2015','2016','2017']].plot(x='Country', kind='bar',stacked=False,width=0.9)






```
